For some of the hurricanes, damage data was not recorded (`"Damages not recorded"`), while the rest are written in the format `"Prefix-B/M"`, where `B` stands for billions (`1000000000`) and `M` stands for millions (`1000000`).
   
   Write a function that returns a new list of updated damages where the recorded data is converted to float values and the missing data is retained as `"Damages not recorded"`.
   
   Test your function with the data stored in `damages`.

````python
# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M',
          '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M',
          '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded',
          '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B',
          '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B',
          '91.6B', '25.1B']

# 1
# Update Recorded Damages
conversion = {"M": 1000000,
             "B": 1000000000}

# test function by updating damages
def update_damages(damages, conversion):
    updated_damages = []
    for damage in damages:
        if damage == 'Damages not recorded':
            updated_damages.append(damage)
        else:
            # Extract the numerical part of the suffix, M or B
            value = float(damage[:-1]) # remove the last character of the suffix and convert to float
            suffix = damage[-1] # last character in the suffix, either M or B
            updated_damages.append(value * conversion[suffix]) # convert using the dictionary
    return updated_damages

updated_damages = update_damages(damages, conversion)
print(updated_damages)
````

3. In addition to organizing the hurricanes in a dictionary with names as the key, you want to be able to organize the hurricanes by year.

   Write a function that converts the current dictionary of hurricanes to a new dictionary, where the keys are years and the values are lists containing a dictionary for each hurricane that occurred in that year.
   
   For example, the key `1932` would yield the value: `[{'Name': 'Bahamas', 'Month': 'September', 'Year': 1932, 'Max Sustained Wind': 160, 'Areas Affected': ['The Bahamas', 'Northeastern United States'], 'Damage': 'Damage not recorded', 'Deaths': 16}, {'Name': 'Cuba II', 'Month': 'November', 'Year': 1932, 'Max Sustained Wind': 175, 'Areas Affected': ['Lesser Antilles', 'Jamaica', 'Cayman Islands', 'Cuba', 'The Bahamas', 'Bermuda'], 'Damage': 40000000.0, 'Deaths': 3103}]`.
   
   Test your function on your hurricane dictionary.

````python
# 3
# Organizing by Year
# create a new dictionary of hurricanes with year and key

def transform_hurricane_dict_by_year(hurricane_dict):
    # Create a new dictionary where the year is the key
    hurricanes_by_year = {}

    # Iterate through the original dictionary
    for hurricane, data in hurricane_dict.items():
        year = data["Year"]  # Extract the year

        # If the year is not already a key, add it with an empty list
        if year not in hurricanes_by_year:
            hurricanes_by_year[year] = []

        # Add the hurricane's data to the list for this year
        hurricanes_by_year[year].append({
            "Name": hurricane,
            "Month": data["Month"],
            "Max Sustained Wind": data["Max Sustained Wind"],
            "Areas Affected": data["Areas Affected"],
            "Damage": data["Damage"],
            "Deaths": data["Deaths"]
        })

    return hurricanes_by_year

# Transform the hurricane data
hurricanes_by_year = transform_hurricane_dict_by_year(hurricane_data)

# Print a sample year to verify the structure
print(hurricanes_by_year[1932])
````

4. You believe that knowing how often each of the areas of the Atlantic are affected by these strong hurricanes is important for making preparations for future hurricanes.

   Write a function that counts how often each area is listed as an affected area of a hurricane. Store and return the results in a dictionary where the keys are the affected areas and the values are counts of how many times the areas were affected.
   
   Test your function on your hurricane dictionary.

````python
# 4
# Counting Damaged Areas
def count_affected_areas(hurricane_dict):
    # Initialize an empty dictionary to store the counts
    area_counts = {}

    # Iterate through the hurricane data
    for hurricane, data in hurricane_dict.items():
        # Get the list of affected areas for this hurricane
        areas = data["Areas Affected"]

        # Count each area
        for area in areas:
            if area not in area_counts:
                area_counts[area] = 1
            else:
                area_counts[area] += 1

    return area_counts

# Test the function on the hurricane dictionary
affected_area_counts = count_affected_areas(hurricane_data)

# Print the results
print(affected_area_counts)
````

5. Write a function that finds the area affected by the most hurricanes, and how often it was hit.

   Test your function on your affected area dictionary.

````python
# 5
# Calculating Maximum Hurricane Count
def most_affected_area(affected_area_counts):
    # Find the area with the maximum count
    most_affected = max(affected_area_counts, key=affected_area_counts.get)
    # Get the count for that area
    count = affected_area_counts[most_affected]
    return most_affected, count

# find most frequently affected area and the number of hurricanes involved in
most_affected, count = most_affected_area(affected_area_counts)
print(f"The area most affected by hurricanes is {most_affected}, with {count} occurrences.")
````

6. Write a function that finds the hurricane that caused the greatest number of deaths, and how many deaths it caused.

   Test your function on your hurricane dictionary.

````python
def deadliest_hurricane(hurricane_dict):
    # Initialize variables to track the deadliest hurricane and its death count
    deadliest = None
    max_deaths = 0

    # Iterate through the hurricane dictionary
    for hurricane, data in hurricane_dict.items():
        # Get the number of deaths for this hurricane
        deaths = data["Deaths"]

        # Update the deadliest hurricane if this one has more deaths
        if deaths > max_deaths:
            deadliest = hurricane
            max_deaths = deaths

    return deadliest, max_deaths

# Test the function on the hurricane dictionary
deadliest, max_deaths = deadliest_hurricane(hurricane_data)

# Print the result
print(f"The deadliest hurricane is {deadliest}, with {max_deaths} deaths.")
````

7. Just as hurricanes are rated by their windspeed, you want to try rating hurricanes based on other metrics.

   Write a function that rates hurricanes on a mortality scale according to the following ratings, where the key is the rating and the value is the upper bound of deaths for that rating.
   
   ```py
   mortality_scale = {0: 0,
   1: 100,
   2: 500,
   3: 1000,
   4: 10000}
   ```
   
   For example, a hurricane with a `1` mortality rating would have resulted in greater than `0` but less than or equal to `100` deaths. A hurricane with a `5` mortality would have resulted in greater than `10000` deaths.
   
   Store the hurricanes in a new dictionary where the keys are the mortaility ratings and the values are lists containing a dictionary for each hurricane that falls into that mortality rating.
   
   Test your function on your hurricane dictionary.

````python
# 7
def rate_hurricanes_by_mortality(hurricane_dict):
    # Mortality scale with ratings and upper bounds
    mortality_scale = {
        0: 0,
        1: 100,
        2: 500,
        3: 1000,
        4: 10000,
        5: float("inf")  # No upper bound for rating 5
    }
    
    # Create a new dictionary to store hurricanes by mortality rating
    rated_hurricanes = {rating: [] for rating in mortality_scale}

    # Iterate through hurricanes
    for hurricane, data in hurricane_dict.items():
        deaths = data["Deaths"]
        
        # Determine the mortality rating for this hurricane
        for rating, upper_bound in mortality_scale.items():
            if deaths <= upper_bound:
                # Add this hurricane to the appropriate rating
                rated_hurricanes[rating].append({
                    "Name": hurricane,
                    "Deaths": deaths,
                    "Year": data["Year"],
                    "Areas Affected": data["Areas Affected"]
                })
                break

    return rated_hurricanes

# Test the function on the hurricane dictionary
rated_hurricanes = rate_hurricanes_by_mortality(hurricane_data)

# Print the results for each rating
for rating, hurricanes in rated_hurricanes.items():
    print(f"Mortality Rating {rating}:")
    for hurricane in hurricanes:
        print(f"  - {hurricane['Name']} ({hurricane['Year']}): {hurricane['Deaths']} deaths")
````

9. Lastly, you want to rate hurricanes according to how much damage they cause.

   Write a function that rates hurricanes on a damage scale according to the following ratings, where the key is the rating and the value is the upper bound of damage for that rating.
   ```py
   damage_scale = {0: 0,
   1: 100000000,
   2: 1000000000,
   3: 10000000000,
   4: 50000000000}
   ```
   
   For example, a hurricane with a `1` damage rating would have resulted in damages greater than `0` USD but less than or equal to `100000000` USD. A hurricane with a `5` damage rating would have resulted in damages greater than `50000000000` USD (talk about a lot of money).
   
   Store the hurricanes in a new dictionary where the keys are damage ratings and the values are lists containing a dictionary for each hurricane that falls into that damage rating.
   
   Test your function on your hurricane dictionary.

````python
# 9
# Rating Hurricanes by Damage
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}

# categorize hurricanes in new dictionary with damage severity as key

def rate_hurricanes_by_damage(hurricane_dict):
    # Damage scale with ratings and upper bounds
    damage_scale = {
        0: 0,
        1: 100000000,       # Up to $100M
        2: 1000000000,      # Up to $1B
        3: 10000000000,     # Up to $10B
        4: 50000000000,     # Up to $50B
        5: float("inf")     # Over $50B
    }
    
    # Create a new dictionary to store hurricanes by damage rating
    rated_hurricanes = {rating: [] for rating in damage_scale}

    # Iterate through hurricanes
    for hurricane, data in hurricane_dict.items():
        damage = data["Damage"]
        
        # Skip hurricanes with unrecorded damage
        if damage == "Damages not recorded":
            continue
        
        # Determine the damage rating for this hurricane
        for rating, upper_bound in damage_scale.items():
            if damage <= upper_bound:
                # Add this hurricane to the appropriate rating
                rated_hurricanes[rating].append({
                    "Name": hurricane,
                    "Damage": damage,
                    "Year": data["Year"],
                    "Areas Affected": data["Areas Affected"]
                })
                break

    return rated_hurricanes

# Test the function on the hurricane dictionary
rated_hurricanes = rate_hurricanes_by_damage(hurricane_data)

# Print the results for each damage rating
for rating, hurricanes in rated_hurricanes.items():
    print(f"Damage Rating {rating}:")
    for hurricane in hurricanes:
        print(f"  - {hurricane['Name']} ({hurricane['Year']}): ${hurricane['Damage']:,.2f} in damages")

````
